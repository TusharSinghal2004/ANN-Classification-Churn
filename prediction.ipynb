```python
import tensorflow as tf
import pandas as pd
from tensorflow.keras.models import load_model
import pickle
import numpy as np
```

LOAD THE TRAINED MODEL , ALL THE PICKLE FILES (SCALED , ONEHOT ENCODE , LABEL)

```python
model = load_model('model.h5')
```

```python
with open('label_encoder_gender.pkl' , 'rb') as file : 
    label_encoder_gender = pickle.load(file)
 
with open('oneHotEncode_geography.pkl' , 'rb') as file : 
    oneHotEncode_geography = pickle.load(file)

with open('scaler.pkl' , 'rb') as file : 
    scaler = pickle.load(file)
```

TIME TO PREDICT USING INPUT DATA

```python
input_data = {
    'CreditScore' : 600,
    'Geography' : 'France' ,
    'Gender' : 'Male' ,
    'Age' : 40 ,
    'Tenure' :  3 ,
    'Balance' : 60000 , 
    'NumOfProducts' : 2 , 
    'HasCrCard' : 1 ,
    'IsActiveMember' : 1 ,
    'EstimatedSalary' : 50000,
    
}
```

```python
## create a data frame for mmy input values 

input_df = pd.DataFrame([input_data])
input_df
```

```python
## Geo encode : 

geo_array = oneHotEncode_geography.transform([[input_data['Geography']]]).toarray()
column_names = oneHotEncode_geography.get_feature_names_out(['Geography'])
geo_encoded_df = pd.DataFrame(geo_array , columns =  column_names)
geo_encoded_df
```

```python
### now we have to concatenate 
input_df = pd.concat([input_df.drop('Geography' , axis= 1) , geo_encoded_df] , axis =1)
input_df
```

```python
## NOW WE USE THE LABEL ENCODER TO ENCODE THE GENDER VALUES INTO IT

input_df['Gender'] = label_encoder_gender.transform(input_df['Gender'])
input_df

```

```python
input_scaled = scaler.transform(input_df)
input_scaled
```

<H4 PREDICT CHURN </H4>

<H4> PREDICT CHURN </H4>

```python
prediction = model.predict(input_scaled)
prediction
```
